```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
# Baixando dados da Petrobras dos últimos 2 anos
df = yf.download("PETR4.SA", period="2y")
print(df.head())
```

```python
# Gráfico do preço de fechamento ao longo do tempo
plt.figure(figsize=(12, 5))
plt.plot(df['Close'], color='blue')
plt.title('Preço de Fechamento - PETR4 (2 anos)')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.grid(True)
plt.show()
```

```python
# Média móvel de 30 e 90 dias
df['MM30'] = df['Close'].rolling(30).mean()
df['MM90'] = df['Close'].rolling(90).mean()

plt.figure(figsize=(12, 5))
plt.plot(df['Close'], label='Preço', alpha=0.5)
plt.plot(df['MM30'], label='Média 30 dias', color='orange')
plt.plot(df['MM90'], label='Média 90 dias', color='red')
plt.title('PETR4 - Preço e Médias Móveis')
plt.xlabel('Data')
plt.ylabel('Preço (R$)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Volume de negociações
volume = df['Volume'].squeeze()

plt.figure(figsize=(12, 4))
plt.bar(volume.index, volume.values, color='purple', alpha=0.5)
plt.title('Volume de Negociações - PETR4')
plt.xlabel('Data')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```
